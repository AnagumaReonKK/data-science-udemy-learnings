# 統計学で使用する用語や数式

## 記述統計

### 正規分布
$\frac{1}{\sqrt{2πσ^2}}exp(-\frac{(x-μ)^2}{2σ^2})$<br>

- 平均μと分散$σ^2$で形状が決まる<br>
$N(μ, σ^2)$

- 68-95-99.7ルール<br>
μ ± σ 約68%<br>
μ ± 2σ 約95%<br>
μ ± 3σ 約99.7%〜100%<br>

### 標準化
$z=\frac{x-\bar{x}}{s}$<br>

### 二項分布
ある観察や試行を複数回行った時に，ある事象が起こる回数が従う確率分布<br>
例）サイコロを3回投げて2がx回でる確率<br>

=サイコロを３回投げて2が1回出る組み合わせ×2が1回出る確率×2が2回でない確率<br>

2がn回中x回でる組み合わせは $_{n}C_{x} = \frac{n!}{x!(n-x)}$<br>

これに2が出る確率をかける<br>

ある事象が起こる確率pの試行をn回実施してその事象がx回起こる確率(P(x))は<br>
$P(x) = _{n}C_{x}p^xq^{n-x} = \frac{n!}{x!(n-x)}p^xq^{n-x}$<br>

ただしq = 1 - p

### 二項分布と正規分布
二項分布のnを無限大にした分布は正規分布に近似する<br>

平均np、分散npqの正規分布に近似できる<br>

例）サイコロを100回投げて2がx回出る確率<br>
100(=n)×$\frac{1}{6}$(=p)→17回が最も確率が高い

## 推測統計

- 統計量: 分布の特徴を表す値 (平均や標準偏差など)
- 標本統計量: 標本から計算された統計量を標本統計量という
- 母数: 母集団の統計量
- 標本: 基本は母集団からの無作為抽出で作成されたもの
- 推定量: 母数の推定に使う標本統計量
- 不偏性: 推定量が”平均的に”母数と一致する
- 標本分布: 標本統計量が従う確率分布

### 標本分布
母集団からランダムに標本を取得する際の標本の背後にある分布のこと<br>
ただし、標本分布は実際には存在しない。

### 不偏性
推定量が”平均的に”母数と一致する場合その推定量は「不偏性がある」という<br>
推定量として適しているかどうかを判断する基準は不偏性があるか？<br>


分散 = $s^2 = \frac{1}{n - 1}Σ^n_{i=1}(x_i - \bar{x})^2$<br>
不偏分散 = $s'^2 = \frac{1}{n - 1}Σ^n_{i=1}(x_i - \bar{x})^2$<br>

## 区間推定
- 母数の値を区間を持って推定すること
- 区間と共に確度をつける(信頼区間)

例）95%の信頼区間とは100回に5回は間違える

## 比率の区間推定
<b>二項分布を用いる</b><br>
$P(x) = _{n}C_{x}p^xq^{n-x} = \frac{n!}{x!(n-x)}p^xq^{n-x}$<br>
標本の数 = n<br>
ある事象が起こる確率 = p<br>
ある事象が起こらない確率 = q<br>

母比率 = p(π)<br>

<b>xの標本分布</b><br>
標本平均 = np<br>
標本分散 = npq<br>
標本標準偏差 $\sqrt{npq}$<br>

<b>比率($\frac{x}{n}$)の標本分布</b><br>
標本比率の平均 = p<br>
標本比率の分散 = $\frac{pq}{n}$<br>
標本比率の標準偏差 $\sqrt{\frac{pq}{n}}$<br>
※標本比率の平均は当然母集団比率p(π)と一致する<br>

<b>標本分布の標準化</b><br>
$z=\frac{\frac{x}{n}-p}{\sqrt{\frac{pq}{n}}}$<br>

-1.96 < z < 1.96<br>

$(n+3.84)p^2 - (2x + 3.84)p + \frac{x^2}{n} < 0$

## 平均値の区間推定
母分散 = $σ^2$<br>
母標準偏差 = $σ$<br>
母平均 = μ<br>

標本平均の分散 = $\frac{σ^2}{n}$(平均の標本分布の分散は母分散をnで割った値になる)<br>
標本平均の標準偏差(標準誤差:SE) = $\sqrt{\frac{σ^2}{n}}$<br>
標本平均($\bar{x}$) ≒ μ<br>
(標本平均の平均は母平均と同じ値になる)<br>
標本平均の平均は標本平均の背後にある標本分布から値を一つ取り出したもの

### z分布
$z = \frac{\bar{x} - μ}{SE} = \frac{\bar{x} - μ}{\frac{σ}{\sqrt{n}}}$<br>

標準正規分布が95%区間でとりうる値<br>
-1.96 < z < 1.96<br>

$\bar{x}-1.96\frac{σ}{\sqrt{n}} < μ < \bar{x}+1.96\frac{σ}{\sqrt{n}}$

### t分布
母標準偏差(σ)の値はわからないので不偏分散($s'^2$)の平方根($s'$)で代用する<br>
$t = \frac{\bar{x} - μ}{\frac{s'}{\sqrt{n}}}$<br>

自由度(degree of freedom) = n-1<br>
$\bar{x}-自由度\frac{s'}{\sqrt{n}} < μ < \bar{x}+自由度\frac{s'}{\sqrt{n}}$<br>

自由度2の場合<br>
$\bar{x}-4.30\frac{s'}{\sqrt{n}} < μ < \bar{x}+4.30\frac{s'}{\sqrt{n}}$<br>

### 仮説
帰無仮説<br>
統計的に差があると言えれば棄却する言えなければ採択する<br>

対立仮説<br>
帰無仮説を棄却できれば対立仮説が成立する<br>

差があるということを証明するのは大変なので、差がないという仮説を立て、否定することで差があることを証明する

<b>仮説検定の流れ<b><br>
1. 帰無仮説と対立仮説を立てる
2. 帰無仮説が正しいと仮定のもと，標本観察を行う(標本統計量を計算する)
3. 標本観察の結果，帰無仮説を否定できるかを確認する(否定した場合，対立仮説が成立)
